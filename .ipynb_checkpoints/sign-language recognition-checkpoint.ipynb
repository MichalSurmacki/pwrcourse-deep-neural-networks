```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_set = pd.read_csv('./data/sign_mnist_train/sign_mnist_train.csv')
train_set.head(10)
```

```python
test_set = pd.read_csv('./data/sign_mnist_test/sign_mnist_test.csv')
test_set.head(10)
```

```python
#Tłumaczenie etykiet na litery alfabetu migowego
train_labels = train_set['label'].values
train_set.drop('label', axis = 1, inplace = True)
unique_labels = np.unique(np.array(train_labels))
letters = ['a','b','c','d','e','f','g','h','i','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y']
letter_label_dictionary = dict(zip(unique_labels, letters))
print("etykieta - litera")
for i in letter_label_dictionary:
    print(i, '-', letter_label_dictionary[i])
```

```python
plt.figure(figsize = (18,8))
sns.countplot(x = train_labels)
```

```python
images = train_set.values
images = np.array([np.reshape(i, (28, 28)) for i in images])
images = np.array([i.flatten() for i in images])
```

```python
from sklearn.preprocessing import LabelBinarizer

label_binarizer = LabelBinarizer()
train_labels = label_binarizer.fit_transform(train_labels)
```
